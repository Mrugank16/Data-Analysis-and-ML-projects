```python
import pandas as pd
```

```python
dataset = pd.read_excel('drive/My Drive/Internshala Assignment/Sunbase/customer_churn_large_dataset.xlsx')
dataset
```

## Data Preprocessing

```python
dataset = dataset.drop(['Name','CustomerID'],axis=1)
dataset
```

```python
dataset.info()
```

```python
from sklearn.preprocessing import LabelEncoder,OneHotEncoder

labelencoder_X = LabelEncoder()
dataset['Gender'] = labelencoder_X.fit_transform(dataset['Gender'])
dataset
```

```python
from sklearn.preprocessing import LabelEncoder,OneHotEncoder

labelencoder_X = LabelEncoder()
dataset['Location'] = labelencoder_X.fit_transform(dataset['Location'])
dataset
```

```python
X = dataset.iloc[:,:-1]
y = dataset.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split

X_train , X_test , y_train , y_test = train_test_split(X ,y , test_size=0.2 , random_state=0)
```

## Feature Engineering

```python
from sklearn.preprocessing import StandardScaler

sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)
```

## Model Building

```python
from xgboost import XGBClassifier
classifier = XGBClassifier()
classifier.fit(X_train , y_train)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
y_pred = classifier.predict(X_test)
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```

## Model Optimization

```python
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator = classifier, X = X_train, y = y_train, cv = 10)
print("Accuracy: {:.2f} %".format(accuracies.mean()*100))
print("Standard Deviation: {:.2f} %".format(accuracies.std()*100))
```
